```python
day = int(input("enter the day number: "))
if(day==1):
    print("mon")
elif(day==2):
    print("tues")
elif(day==3):
    print("Wed")
elif(day==4):
    print("thur")
elif(day==5):
    print("Fri")
elif(day==6):
    print("Sat")
else:
    print("Sunday")
```

```python
day = int(input("Enter the day number: "))

match day:
    case 1:
        print("mon")
    case 2:
        print("tues")
    case 3:
        print("wed")
    case _:
        print("invalid number")
```

```python
# wap to check given char is vowel or consonent.

char = "b"

match char:
    case 'a'|'e'|'i'|'o'|'u':
        print("vowel hai")
    case _:
        print("consonent")
```
