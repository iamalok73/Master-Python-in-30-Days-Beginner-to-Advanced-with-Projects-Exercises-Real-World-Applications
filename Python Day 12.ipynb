```python
class ATM:
    def __init__(self,amount):
        self._amount = amount

    def Check_Balance(self):
        print(f"Your Current Balance is {self._amount}")

    def Deposite(self,amount):
        if amount >0:
            self._amount +=amount
            print(f"You Deposite amount is {amount}")
            print(f"Your Current Balance is {self._amount}")
        else:
            print('Deposite +ve positive value.')
            
    def Withdraw(self,amount):
        if amount <= 0:
            print("Please enter +ve Amount.")
        elif(amount>self._amount):
            print("Ensuffiencent balance in your acount.")
        else:
            self._amount -= amount
            print(f"ypour withdraw amount is {amount}")
            print(f"And ypour current balance is {self._amount}")
```

```python
atm = ATM(200)
```

```python
atm.Check_Balance()
```

```python
atm.Withdraw(10)
```

```python
atm.Deposite(-210)
```
