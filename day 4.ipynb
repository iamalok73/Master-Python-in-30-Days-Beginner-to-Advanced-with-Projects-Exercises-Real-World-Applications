```python
num = int(input("enter the number: "))
if(num>0):
    print("this number is possitive")
    if(num%2==0):
        print("even")
    else:
        print("odd")
else:
    print("this number is negative")
    if(num%2==0):
        print("even")

    else:
        print("odd")
```

```python
# wap to find sum of 10 natural numbe.

num = int(input("enter the number: "))

sum = 0 

for i in range(1,num+1):
    sum+=i
print(f"The Total sum is {sum}.")
```

```python
# wap to find multiple  of 10 natural numbe.

num = int(input("enter the number: "))

prod = 1

for i in range(1,num+1):
    prod*=i
print(f"The Total sum is {prod}.")
```
