### While Loops

- it is used when we want to repeat again and again untill oure condtion not false.
- it work condition based

```python
# wap to print 5 number start from 1.

start = 1
end = 5

while start<=end:
    print(start)
    start+=1
```

```python
# wap to print 10 natural number using while loop.

start = 1
end = 10

while start<=end:
    print(start)
    start+=1
```

```python
# wap to print 10 natural number in reverse order.

start = 10
end = 1

while start>=end:
    print(start)
    start-=1
```

```python
# wap to print table and the input take from the user using loop.
num = int(input("enter the number: "))

for i in range(1,10+1):
    print(f"{num} * {i} = {i*num}.")
```

```python
num = int(input("enter the number: "))
s = 1
e = 10
while  s<=e:
    print(f"{num} * {s} = {s*num}")
    s+=1
```
