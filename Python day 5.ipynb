```python
# wap to find sum of n numbers.

num = int(input("enter the number: "))
total = 0
for i in range(1,num+1):
    total+=i
print(f"The Total Amount is : {total}")
print("The Total Amount is : ",total)
```

```python
# wap to store only name who have start with a or A.
# [Amit, raju ,sumit ,Atul,ashish]
# o/p = Amit,Atul,ashish

name = []
l = ["Amit", "Raju","Sumit", "Atul","ashish"]

for i in l:
    if i.capitalize().startswith(('a',"A")):
        name.append(i)

print(f"my new list is = {name}")
```

```python
# wap to print only those name who have more than 5 latter in name.
l = ["Amit", "Raju","Sumit", "Atul","ashish"]

for i in l:
    if len(i)>=5:
        print(i)
```

```python
## remove the duplicate names from the list

l = ["Amit", "Raju","Sumit", "Atul","ashish","ashish","ashish","Amit","Raju"]
# o/p = ["Amit", "Raju","Sumit", "Atul","ashish"]
new_l = []

for i in l:
    if i not in new_l:
        new_l.append(i)
print(new_l)       
```

```python
### find the longest name in list.
l = ["Amit", "Raju","Sumit", "Atul","ashish"]
longest = l[0]

for i in l:
    if len(i)>len(longest):
        longest = i
print(longest)
```
