```python
# Match-case
```

```python
# match variable:
#     case value:
#         # block of code
```

```python
day_num = int(input("Enter the day number: "))

match day_num:
    case 1:
        print(f"my day name is Monday")
    case 2:
        print(f"my day name is Tuesday.")
    case 3:
        print(f"my day name is Wednesday.")
    case 4:
        print(f"my day name is Thursday.")
    case 5:
        print(f"my day name is Friday.")
    case 6:
        print(f"my day name is Saterday.")
    case 7:
        print(f"my day name is Sunday.")
    case _:
        print("invalid number")
```

```python
# wap to for if user enter sat, or sunday then weekend otherwise weekday.

day = input("Enter the Day name: ")

match day:
    case "Saterday" | "Sunday":
        print("its Weekend")
    case _:
        print("Weekday")
```

```python
# wap to check user is minor or major

Age = int(input("enter the Age"))

match Age:
    case x  if x >= 18:
        print("major")
    case x  if x < 18:
        print("minor")
```
