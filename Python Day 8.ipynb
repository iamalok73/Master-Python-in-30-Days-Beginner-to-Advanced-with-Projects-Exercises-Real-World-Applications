```python
def fact(n):
    fac = 1
    for i in range(1,n+1):
        fac*=i
    print(f"The Factorial is {fac}")
```

### Control Statement

- distrub the flow of program.
- break
- pass
- continue

```python
for i in range(1,10):
    if(i==5):
        break
    print(i)
```

```python
for i in range(1,10):
    pass
```

```python
for i in range(1,10):
    if(i==5):
        continue
    print(i)
```

### Lambda Function
- Lambda function is a annonymous function.
- it is small function.
- it is one line function.
- use by lambda keyword.

```python
z = lambda a,b :a+b
```

```python
print(z(2,4))
```

```python
n = lambda x : x+10
```

```python
print(n(2))
```

```python
c = lambda x,y,z : x*y*z
```

```python
print(c(2,3,4))
```
