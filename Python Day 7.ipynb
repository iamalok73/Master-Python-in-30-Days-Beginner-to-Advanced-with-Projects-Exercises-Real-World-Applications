```python
# find the sum of digit number.

l = int(input("enter the number"))
print(l)
sum = 0
for i in l:
    sum+=i
print(f"Total Sum of = {sum}.")
```

```python
num = int(input("enter the number"))
sum_digi = 0
for i in str(num):
    sum_digi += int(i)
print(f"Total Sum of = {sum_digi}.")
```

```python
num = 12345678910
suum = 0

while num > 0 :
    suum +=num%10
    num//=10
print(f"Total Sum of thiss number is {suum}.")
```

```python
# # Revere any number

n = 987654321
rev = 0
while n>0:
    rem = n%10
    rev = rev*10 + rem 
    n //=10
print(rev)
```

```python
# range -- 
for j in range(2,3):
    for i in range(1, 11):
        print(f"{j} * {i} = {j*i} ")
```

## Nested loops in python
- when the loop write inside another loops is called nested loop

```python
# wap to print one side trinagel
```

```python
# *
# * *
# * * *
```

```python
for i in range(1,4):
    for j in range(1,i+1):
        print("*" , end = " ")
    print(" ")
```

```python
# reverse 
# * * *
# * *
# *

   
for i in range(1,4):
    for j in range(1,4-i+1):
        print("*" , end = " ")
    print(" ")
```

```python
# triangle 

#     *
#   * * * 
# * * * * *
```

```python
for i in range(1,6):
    for j in range(1,6-i):
        print(" ", end= " ")
    for k in range(1,2*i):
        print("*", end= ' ')
    print(" ")
```

```python
for i in range(1,6):
    for j in range(1,6-i):
        print(" ", end= " ")
    for k in range(1,2*i):
        print("*", end= ' ')
    print(" ")
```

```python
#      *
#    * *
# * *  *


for i in range(4):
    for k in range(i,4):
        print(" ", end= " ")
    for j in range(i+1):
        print("*", end = " ")
    print(" ")
```

```python
# pritn hollow square

# * * * * 
# *     *
# *     *
# * * * *

num = int(input("enter the number : "))
for i in range(num):
    for j in range(num):
        if(i==0 or i==num-1 or j == 0  or j==num-1 ):
            print("*", end = " ")
        else:
            print(" ", end = " ")
    print(" ")
```
