### Operators 

- The Operators are perform to some calculation on the operands is called operators

- In python , there are multiple type of operators.

1. Arithmatic operators.  -> + , - , * , / , % , // , **
2. Assignment Operators.  -> += , -= , *= , /= , %= , //= , **=
3. Comperission / Relational Operators.  ->  > , < , >= , <= , == , != 
4. Logical Operaotrs.   --> And , or , not 
5. Membership Operators.  --> in , not in
6. identity Operators.  -->  is , is not
7. Bitwise Operators.  --> & , ` , ^ , ~ , << , >> 

###  Arithmatic operators.  --> + , - , * , / , % , // , **

```python
a = 20
b = 15

print(a+b)  # 35
print(a-b) # 5
print(a*b)  # 300
print(a/b)  # 1.333


# % moduls -- define remainder values

print(a%b)  # 5

# // floor division -- exaclty ...

print(a//b) # 1


# ** power

print(3**3)
```

```python
# Nested condition 
# when a condition written inside another condition is called nested condition.
```

```python
marks = int(input("Enter the number: "))
if(marks>100):
    print("please enter valid marks")  
elif(marks<=100 and marks >=0): 
    if(marks>=80):
        print("A Grade")
    elif(marks < 80 and marks >=55):
        print("B Grade")
    elif(marks <55 and marks >=33):
        print("C grade")
    else:
        print("Fails")
else:
    print("invalide marks")
```
