- comprehension is a method to short an callucatoin or syntax of sequence.(collection).

### List Comprehenshion

- use to create a list

```python
# [expression for item in iterable if condition ]
```

```python
l= [5,1,3,2,4]
```

```python
# wap to print square of each element in the list
```

```python
li = [ ]

for i in l:
    li.append(i**2)
print(li)
```

```python
result = list(map(lambda a:a**2, l))
print(result)
```

```python
z = [i**2 for i in l ]
print(z)
```

```python
# print only even value
z = [i**2 for i in l if i%2==0 ]
print(z)
print(type(z))
```

### Set Comprehension

```python
s = {i**2 for i in l if i%2==0 }
print(s)
print(type(s))
```

### Tuple Compreshension

```python
s = tuple(i**2 for i in l if i%2==0 )
print(s)
print(type(s))
```

### Dictinary Comprehension

```python
d = {i: i**2 for i in l if i%2==0 }
print(d)
print(type(d))
```

### Nested Comprehension

```python
for i in range(3):
    for j in range(3):
        print(j, end=" ")
    print(" ")
```

```python
re = [ [i for i in range(3)] for j in range(3)]
print(re)
```

```python
# find even odd using compreshnsion from range(1,6)
```

```python
re = ["even" if i %2==0 else "odd" for i in range(1,6)]
print(re)
```

### Flatten Array

- convert nested condition t0 single line.

```python
re = [[1,2],[3,4],[4,5]]
len(re)
```

```python
re_new = []
for i in re:
    for j in i:
        re_new.append(j)

print(re_new)
```

```python
ss = [j for i in re for j in i]
ss
```
