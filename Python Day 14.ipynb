### Hierarchical Inhertance

- When a single class share our property to multiple child class is called Hierarchical Inhertance.

```python
class PaymentGateway:
    def pay(self,amount):
        raise NotImplementedError
```

```python
class UPI(PaymentGateway):
    def pay(self,amount):
        return f"Paid {amount} using UPI."

class NetBanking(PaymentGateway):
    def pay(self,amount):
        return f"Paid {amount} using NetBanking."

class Card(PaymentGateway):
    def pay(self,amount):
        return f"Paid {amount} using Card."
```

```python
upi = UPI()
```

```python
upi.pay(100)
```

```python
net = NetBanking()
```

```python
net.pay(120)
```

```python
c =Card()
```

```python
c.pay(500)
```

```python
payment = [UPI(),NetBanking(),Card()]
```

```python
for p in payment:
    a = 50
    print(p.pay(a))
```

```python
### Method overriding +super keyword
```

```python
class Vehical:
    def max_speed(self):
        return 150
```

```python
class Car(Vehical):
    def max_speed(self):
        base_speed = super().max_speed()
        return base_speed+100
```

```python
c = Car()
```

```python
c.max_speed()
```

### Polymorphism

- it means many form.
- In python , the same method name can behave differntly depends on. 

```python
class Test:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
```

```python
t= Test()
```

```python
class Dog:
    def sound(self):
        print("Barks")

class Cat:
    def sound(self):
        print("meow")
        
```

```python
def animal_sound(n):
    n.sound()
```

```python
animal_sound(Dog())
```

```python
animal_sound(Cat())
```

```python
class Book:
    def __init__(self,pages):
        self.pages = pages

    def __add__(self,other):
        return self.pages+other.pages
```

```python
b_1 =  Book(100)
```

```python
b_2 = Book(200)
```

```python
print(b_1+b_2)
```

```python
def Fact(n):
    if n==1:
        return 1
    else:
        return Fact(n-1)*n
```
