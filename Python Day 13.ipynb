### INHERITENCE

- When a class share our properties to another class , is called Inheritence.
- When a Child class aquries Parents class Property.
- which class share share our property to another class is called Parents class.
- Which class aquaries to another class property is called Child Class.

```python
class Parents:
    def Home(self):
        print("this is Parents class home.")
    def Money(self):
        print("this is Father Money.")
```

```python
class Child(Parents):
    def school(self):
        print("This is child class")
    def Toys(self):
        print("This is Child toys.")
```

```python
P = Parents()
```

```python
P.Home()
P.Money()
```

```python
C = Child()
```

```python
C.school()
```

```python
C.Toys()
```

```python
C.Home()
```

### Single Level Inheritence

```python
class Animal:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def speak(self):
        print(f"{self.name} make noise.")
```

```python
class Dog(Animal):
    def __init__(self,name,age,bread):
        super().__init__(name,age)
        self.bread = bread
        
    def Bark(self):
        print(f"The dog name is {self.name} and bread is {self.bread} and Age is {self.age}, Barks.")

        
```

```python
d = Dog("Kallu",2,"indian")
```

```python
d.name
```

```python
d.bread
```

```python
d.Bark()
```

```python
d.speak()
```

```python
d.age
```

### Multi level Inheritence

```python
class Company:
    def company_policy(self):
        return "9 hours working."
```

```python
class Employee(Company):
    def __init__(self,emp_id,salary):
        self.emp_id = emp_id
        self.salary = salary
    def annual_salary(self):
        return self.salary*12
```

```python
class Manager(Employee):
    def __init__(self,emp_id, salary, bonus):
        super().__init__(emp_id,salary)
        self.bonus = bonus
    def total_income(self):
        return self.annual_salary()+self.bonus
    
```

```python
mgr = Manager(112,80000,50000)
```

```python
mgr.bonus
```

```python
mgr.salary
```

```python
mgr.emp_id
```

```python
mgr.company_policy()
```

```python
mgr.annual_salary()
```

```python
mgr.total_income()
```

### Multiple Inheritence

```python
### Authentication
```

```python
# login 
# Permission

# system Secure
```

```python
class Login:
    def authenticate(self,username,password):
        return username == "Admin@123" and password == "1234"
```

```python
class Permission:
    def has_access(self,role):
        return role in ["Admin","Manager"]
```

```python
class Secure_System(Login,Permission):
    def access_system(self,username,password,role):
        if self.authenticate(username,password) and self.has_access(role):
            return "Access Granted"
        else:
            return "Access Denied"
```

```python
sys = Secure_System()
```

```python
sys.access_system("Admin@123","1234","Admin")
```

```python
sys.access_system("Admin@1234",1234,"Admin")
```
