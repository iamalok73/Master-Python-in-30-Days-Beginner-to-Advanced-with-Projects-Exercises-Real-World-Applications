### Set Data type

- set are store only unique element.
- set are un ordered data type.
- set are mutable data type.
- set is not follow nested set.
- set are denoted by {}.

```python
s = {1,1,1,2,"Amit",-1,2.1,1.5,"3"}
print(s)
print(type(s))
```
